```python
#배열분할 split
x=[1, 2, 3, 77, 77, 3, 2, 1]
x1, x2, x3 = np.split(x,[3, 5])
print(x1,x2,x3)
```

```python
#배열분할 vsplit
grid = np.arange(16).reshape((4,4))
upper, lower=np.vsplit(grid,[2])
print(upper)
print(lower)
```

```python
#배열분할 hsplit
grid = np.arange(16).reshape((4,4))
left, right=np.hsplit(grid,[2])
print(left)
print(right)
```
