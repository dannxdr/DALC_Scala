#### 딕셔너리 구조
- 일련의 임의의 값에 임의의 키를 매핑
- 타입이 지정된 키를 일련의 타입이 지정된 값에 매핑하는 구조

```python
population_dict = {'California' : 38332621,'Texas':26448393,'New York' : 19651127,'Florida' : 19552860,'Illinois':12882135}
population = pd.Series(population_dict)
population
```

```python
#인덱스가 정렬된 키에서 추출되는 경우 값을 불러옴
population['California']
```

```python
#슬라이싱 같이 배열 스타일의 연산 지원
population['California' : 'Illinois']
```

```python
#리스트 혹은 넘파이 배열, 인덱스 생략가능
pd.Series([2,4,6])
```

####DataFrame
- 행 인덱스 : 시리즈 인덱스와 동일
- 열 인덱스 : 열 이름
- 데이터프레임 : 행 인덱스와 열 이름을 갖는 행렬, 시리즈 객체의 연속


```python
area_dict = {'California' : 423967,'Texas': 696565,'New York' : 1419297,'Florida' : 170312,'Illinois':149995}
area = pd.Series(area_dict)
area
```

```python
states = pd.DataFrame({'population':population,'area':area})
states
```

```python
states.index
```

```python
states.columns
```

```python
states['area']
```

```python
#단일 시리즈로 데이터 프레임 구성 가능
pd.DataFrame(population,columns=['population'])
```

```python
pd.DataFrame(population,index=['Texas'],columns=['population'])
```
